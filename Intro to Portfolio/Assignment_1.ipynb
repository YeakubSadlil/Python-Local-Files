**Define annualized return and annualized volatility functions**

```python
def annualized_ret(r,periods_per_year):
    compounded_growth = (1+r).prod()
    return compounded_growth**(periods_per_year/r.shape[0]) -1

def annualized_vol(r,periods_per_year):
    return r.std()*(periods_per_year**0.5)
```

### Calculate annualized return, annualized volatility, return_on_risk_ratio based on the closing price

```python
sector_stats = {}
sectors = [tech, healthcare, consumer, finance]
sect = ["tech", "healthcare", "consumer", "finance"]

# Data was taken month wise from 2020-01-01 to 2023-01-01

for i,s in enumerate(sectors):
    stocks = yf.download(s, start='2020-01-01', end='2023-01-01', interval="1mo")['Close']  
    returns = stocks.pct_change().dropna()
    
    ann_ret = annualized_ret(returns, 12)
    ann_vol = annualized_vol(returns, 12)
    return_on_risk_ratio = ann_ret/ann_vol
    sector_stats[sect[i]] = return_on_risk_ratio
```

**Plot all the stocks based on Return-to-Risk Ratio**

```python
all_stocks_ret_risk_ratio = [ratio for ratios in sector_stats.values() for ratio in ratios]
all_stocks = np.concatenate(sectors)

data = {'Stock': all_stocks, 'Return-to-Risk Ratio': all_stocks_ret_risk_ratio}
df = pd.DataFrame(data)
sorted_df = df.sort_values(by='Return-to-Risk Ratio', ascending=False)

plt.figure(figsize=(10, 6))
plt.bar(sorted_df['Stock'], sorted_df['Return-to-Risk Ratio'], color='b')
plt.xlabel('Stocks')
plt.ylabel('Return-to-Risk Ratio')
plt.title('Return-to-Risk Ratio for Each Stock')
plt.xticks(rotation=40, ha='right')
plt.show()
```

**Define functions to find effiecent frontier**

```python
from scipy.optimize import minimize
def portfolio_return(weights, returns):
    return weights.T @ returns


def portfolio_vol(weights, covmat):
    return (weights.T @ covmat @ weights)**0.5

def minimize_vol(target_return, er, cov):
    n = er.shape[0]
    init_guess = np.repeat(1/n, n)
    bounds = ((0.0, 1.0),) * n 

    weights_sum_to_1 = {'type': 'eq',
                        'fun': lambda weights: np.sum(weights) - 1
    }
    return_is_target = {'type': 'eq',
                        'args': (er,),
                        'fun': lambda weights, er: target_return - portfolio_return(weights,er)
    }
    weights = minimize(portfolio_vol, init_guess,
                       args=(cov,), method='SLSQP',
                       options={'disp': False},
                       constraints=(weights_sum_to_1,return_is_target),
                       bounds=bounds)
    return weights.x


def optimal_weights(n_points, er, cov):
    target_rs = np.linspace(er.min(), er.max(), n_points)
    weights = [minimize_vol(target_return, er, cov) for target_return in target_rs]
    return weights


def plot_ef(n_points, er, cov):
    weights = optimal_weights(n_points, er, cov)
    rets = [portfolio_return(w, er) for w in weights]
    vols = [portfolio_vol(w, cov) for w in weights]
    ef = pd.DataFrame({
        "Returns": rets, 
        "Volatility": vols
    })
    return ef.plot.line(x="Volatility", y="Returns", style='.-', legend=False)
```
