### Rolling Correlation - multiindexes and .groupby

```python
ts_corr = ind_return.rolling(window=36).corr()
ts_corr.tail()
```

```python
ts_corr.index.names = ["date","industry"]
ts_corr.tail()
```

```python
ind_tr36corr = ts_corr.groupby(level="date").apply(lambda corrmat: corrmat.values.mean())
ind_tr36corr.tail()
```

```python
tmi_tr36rets.plot(label="trailing_36 rets",legend=True,figsize=(12,8))
ind_tr36corr.plot(label="ind_tr36corr",legend=True,secondary_y=True)
```

```python
tmi_tr36rets["2007":].plot(label="trailing_36 rets",legend=True,figsize=(12,8))
ind_tr36corr["2007":].plot(label="ind_tr36corr",legend=True,secondary_y=True)
```

The concept behind diversification is to spread investments across different assets or asset classes to reduce the overall risk of a portfolio. The idea is that when some investments perform poorly, others may perform well, helping to mitigate losses. This strategy is based on the assumption that the returns of different assets are not perfectly correlated.

Correlation measures the degree to which two assets move in relation to each other. If assets are strongly positively correlated, it means they tend to move in the same direction. In contrast, if they are strongly negatively correlated, they move in opposite directions.

When assets are strongly positively correlated, it implies that they are likely to move in the same direction during market fluctuations. If one asset performs poorly, the other is also likely to perform poorly. In such a scenario, the risk-reducing benefits of diversification are limited because the portfolio is exposed to similar risks across different assets. Diversification works best when assets have a lower or negative correlation, as this means they are less likely to move in tandem.

Investors seek assets with low or negative correlations to achieve effective diversification. By combining assets with diverse risk-return profiles and low correlations, investors aim to create a portfolio that can perform well across different market conditions, reducing overall portfolio risk.
